```python
###########################################
#
# tkinter 필요 클래스 모음
#
###########################################
import tkinter as tk

class Frame_ListBox:
    def __init__(self,window,**kwargs):
        self.frame = tk.Frame(window)
        self.frame.pack()

        self.scrollbar = tk.Scrollbar(self.frame)
        self.scrollbar.pack(side='right', fill='y')

        self.listbox = tk.Listbox(self.frame,selectmode='extended', height = 10, yscrollcommand=self.scrollbar.set)
        self.listbox.pack()

        self.scrollbar.config(command=self.listbox.yview)

    # 순차적 리스트 생산
    def addList(self,item):
        self.listbox.insert(self.listbox.size(),item)

###########################################
#
# 데이터 프레임 파일 불러오기
#
###########################################
import pandas as pd

###########################################
#
# AI Algorithm 학습 모음
#
###########################################
# sklearn 에 적재되어 있는 알고리즘

import pandas as pd
import numpy as np        

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier

# 설치가 필요한 앙상블 알고리즘 
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
#from catboost import CatBoostClassifier

class AIClasssifierModel:
    def __init__(self, df,**kwargs):
        pass

class AIRegressionModel:
    def __init__(self, df, **kwargs):
        pass

###########################################
#
# 메인
#
###########################################
import sklearn.datasets as sk_data

window = tk.Tk()
window.title('Tool')
window.geometry('640x400+100+100')
window.resizable(True,True)

right_frame = tk.Frame(window)
right_frame.pack(side='right',fill='both',expand=True,padx=5,pady=5)

left_frame = tk.Frame(window)
left_frame.pack(side='left',fill='both',expand=True,padx=5,pady=5)

# AI algorithm listbox
ai_listbox = Frame_ListBox(right_frame)
ai_listbox.addList('test')


sample_data_listbox = Frame_ListBox(right_frame)

for sample in sk_data.__all__:
    sample_data_listbox.addList(sample)

window.mainloop()
```
