```python
import tkinter as tk
from tkinter import ttk
import pandas as pd

class DataFrameEditor:
    def __init__(self, root):
        self.root = root
        self.root.title("DataFrame 편집기")

        # 빈 DataFrame 생성
        self.df = pd.DataFrame()

        # Treeview 생성
        self.tree = ttk.Treeview(self.root, columns=[], show="headings")
        self.tree.bind('<ButtonRelease-1>', self.on_treeview_click)
        self.tree.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")

        # Scrollbar 생성
        scrollbar = ttk.Scrollbar(self.root, orient="vertical", command=self.tree.yview)
        self.tree.configure(yscroll=scrollbar.set)
        scrollbar.grid(row=0, column=1, sticky="ns")

        # 행, 열 추가 및 삭제 버튼
        add_row_button = ttk.Button(self.root, text="행 추가", command=self.add_row)
        add_col_button = ttk.Button(self.root, text="열 추가", command=self.add_column)
        delete_row_button = ttk.Button(self.root, text="행 삭제", command=self.delete_row)
        delete_col_button = ttk.Button(self.root, text="열 삭제", command=self.delete_column)

        add_row_button.grid(row=1, column=0, pady=5, sticky="ew")
        add_col_button.grid(row=2, column=0, pady=5, sticky="ew")
        delete_row_button.grid(row=3, column=0, pady=5, sticky="ew")
        delete_col_button.grid(row=4, column=0, pady=5, sticky="ew")

    def on_treeview_click(self, event):
        # Treeview의 열 제목을 클릭했을 때, 해당 열을 선택하도록 함
        col_id = self.tree.identify_column(event.x)
        if col_id:
            col_index = col_id.split("#")[-1]
            self.tree.selection_set((col_index,))

    def add_row(self):
        # 빈 행 추가
        self.df = self.df.append({}, ignore_index=True)
        self.update_treeview()

    def add_column(self):
        # 사용자로부터 열 이름 입력 받음
        col_name = tk.simpledialog.askstring("열 추가", "추가할 열의 이름을 입력하세요:")
        if col_name and col_name not in self.df.columns:
            # 빈 열 추가
            self.df[col_name] = ""
            self.update_treeview()

    def delete_row(self):
        # 선택된 행 삭제
        selected_items = self.tree.selection()
        for item in selected_items:
            row_index = int(item) - 1
            self.df = self.df.drop(index=row_index)
        self.update_treeview()

    def delete_column(self):
        # 선택된 열 삭제
        selected_items = self.tree.selection()
        for item in selected_items:
            col_index = int(item) - 1
            col_name = self.df.columns[col_index]
            self.df = self.df.drop(columns=col_name)
        self.update_treeview()

    def update_treeview(self):
        # 기존 데이터 삭제
        for row in self.tree.get_children():
            self.tree.delete(row)

        # DataFrame 데이터 추가
        for i, row in self.df.iterrows():
            self.tree.insert("", "end", values=list(row))

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFrameEditor(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk, filedialog
import pandas as pd

class DataFrameEditor:
    def __init__(self, root):
        self.root = root
        self.root.title("DataFrame 편집기")

        # DataFrame을 저장할 변수
        self.df = pd.DataFrame()

        # Treeview 생성
        self.tree = ttk.Treeview(self.root, columns=[], show="headings")
        self.tree.bind('<ButtonRelease-1>', self.on_treeview_click)
        self.tree.grid(row=0, column=0, padx=10, pady=10, sticky="nsew")

        # Scrollbar 생성
        scrollbar = ttk.Scrollbar(self.root, orient="vertical", command=self.tree.yview)
        self.tree.configure(yscroll=scrollbar.set)
        scrollbar.grid(row=0, column=1, sticky="ns")

        # 파일 타입 선택 라디오 버튼
        self.file_type_var = tk.StringVar(value="csv")

        csv_radio = tk.Radiobutton(self.root, text="CSV", variable=self.file_type_var, value="csv")
        excel_radio = tk.Radiobutton(self.root, text="Excel (XLSX)", variable=self.file_type_var, value="xlsx")
        json_radio = tk.Radiobutton(self.root, text="JSON", variable=self.file_type_var, value="json")

        csv_radio.grid(row=1, column=0, padx=10, pady=5, sticky="w")
        excel_radio.grid(row=1, column=1, padx=10, pady=5, sticky="w")
        json_radio.grid(row=1, column=2, padx=10, pady=5, sticky="w")

        # 파일 선택 버튼
        select_file_button = tk.Button(self.root, text="파일 선택", command=self.load_file)
        select_file_button.grid(row=2, column=0, columnspan=3, pady=10)

        # 행, 열 추가 및 삭제 버튼
        add_row_button = tk.Button(self.root, text="행 추가", command=self.add_row)
        add_col_button = tk.Button(self.root, text="열 추가", command=self.add_column)
        delete_row_button = tk.Button(self.root, text="행 삭제", command=self.delete_row)
        delete_col_button = tk.Button(self.root, text="열 삭제", command=self.delete_column)

        add_row_button.grid(row=3, column=0, pady=5, sticky="ew")
        add_col_button.grid(row=3, column=1, pady=5, sticky="ew")
        delete_row_button.grid(row=3, column=2, pady=5, sticky="ew")
        delete_col_button.grid(row=3, column=3, pady=5, sticky="ew")

    def on_treeview_click(self, event):
        # Treeview의 열 제목을 클릭했을 때, 해당 열을 선택하도록 함
        col_id = self.tree.identify_column(event.x)
        if col_id:
            col_index = col_id.split("#")[-1]
            self.tree.selection_set((col_index,))

    def load_file(self):
        file_type = self.file_type_var.get()

        # 파일 선택 대화상자 열기
        file_path = filedialog.askopenfilename(filetypes=[(f"{file_type.upper()} files", f"*.{file_type}")])

        if file_path:
            try:
                # 선택한 파일을 DataFrame으로 읽어오기
                if file_type == "csv":
                    self.df = pd.read_csv(file_path)
                elif file_type == "xlsx":
                    self.df = pd.read_excel(file_path)
                elif file_type == "json":
                    self.df = pd.read_json(file_path)

                # 읽어온 DataFrame을 Treeview에 출력
                self.update_treeview()

            except Exception as e:
                tk.messagebox.showerror("에러", f"파일을 읽어오는 중 에러가 발생했습니다:\n{str(e)}")

    def add_row(self):
        # 빈 행 추가
        self.df = self.df.append({}, ignore_index=True)
        self.update_treeview()

    def add_column(self):
        # 사용자로부터 열 이름 입력 받음
        col_name = tk.simpledialog.askstring("열 추가", "추가할 열의 이름을 입력하세요:")
        if col_name and col_name not in self.df.columns:
            # 빈 열 추가
            self.df[col_name] = ""
            self.update_treeview()

    def delete_row(self):
        # 선택된 행 삭제
        selected_items = self.tree.selection()
        for item in selected_items:
            row_index = int(item) - 1
            self.df = self.df.drop(index=row_index)
        self.update_treeview()

    def delete_column(self):
        # 선택된 열 삭제
        selected_items = self.tree.selection()
        for item in selected_items:
            col_index = int(item) - 1
            col_name = self.df.columns[col_index]
            self.df = self.df.drop(columns=col_name)
        self.update_treeview()

    def update_treeview(self):
        # 기존 데이터 삭제
        for row in self.tree.get_children():
            self.tree.delete(row)

        # DataFrame 데이터 추가
        for i, row in self.df.iterrows():
            self.tree.insert("", "end", values=list(row))

if __name__ == "__main__":
    root = tk.Tk()
    app = DataFrameEditor(root)
    root.mainloop()
```
